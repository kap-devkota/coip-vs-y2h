```python
import pandas as pd
```

```python
gene2pub = pd.read_csv("gene-pubmed-data/gene2pubmed", sep = "\t")
gene2pub
```

```python
geneinfo = pd.read_csv("gene-pubmed-data/gene_info", sep = "\t")
geneinfo
```

```python
genehuman = geneinfo.loc[geneinfo.iloc[:, 0].apply(lambda x : x == 9606), ["GeneID", "Symbol"]]
```

```python
g2pubhuman = gene2pub.loc[gene2pub.iloc[:, 0].apply(lambda x: x == 9606), ["GeneID", "PubMed_ID"]]
```

```python
g2pubhuman = g2pubhuman.drop_duplicates()
g2pubhuman
```

```python
g2idcounts = g2pubhuman.groupby("GeneID", as_index=False).count()
g2idcounts.sort_values(["PubMed_ID"],ascending=False) 
```

```python
combined = pd.merge(g2idcounts, genehuman, how="inner", on="GeneID")
combined.columns = ["GeneID", "#PubMed", "Symbol"]
combined = combined.sort_values(["#PubMed"], ascending=False)
```

```python
combined.to_csv("data/pubmed_counts.tsv", sep = "\t", index = None)
```
