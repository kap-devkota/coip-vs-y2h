```python
import torch
import torch.nn as nn


class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.x = nn.Parameter(torch.randn(5, 5), requires_grad = True)
        self.y = nn.Parameter(torch.randn(5, 5), requires_grad = True)
    
    def forward(self, z):
        return torch.matmul(z, self.x) * (torch.matmul(z, self.y) + 1)
    
model = Model()
```

```python
model.x.requires_grad = False
model.x.requires_grad
opt = torch.optim.Adam(model.parameters(), lr = 1e-2)
```

```python
x = torch.randn(10, 5)
y = torch.randn(10, 5)
lossf = torch.nn.MSELoss()

for i in range(10):
    opt.zero_grad()
    output = model(x)
    loss = lossf(y, output)
    loss.backward()
    print(model.y.grad)
    opt.step()
    print(f"\n\nEpoch {i}")
```
