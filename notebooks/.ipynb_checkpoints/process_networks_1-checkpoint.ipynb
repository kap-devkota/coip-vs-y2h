```python
mapping = pd.read_csv("../data/rsingh/HUMAN_9606_idmapping.dat", sep = "\t", header = None)
mapping = mapping[mapping[1] == "GeneID"][[0, 2]].reset_index(drop = True)
mapping[0] = "uniprotkb:" + mapping[0]
mapping.columns = ["uniprot", "entrez"]
mapping.to_csv("../data/idmap/uprot_entrez_new.tsv", sep = "\t", index = None)
```

```python
uniprot_mapping = {f"uniprotkb:{f}" for f in mapping[0]}
uniprot_mapping
```

```python
# Get the CoIP network
coip = pd.read_csv("../data/rsingh/human_intact_my_annotations.csv")
coip = coip[coip["Category"] == "CoIP"][["id1", "id2", "score"]]
coip.loc[coip["id1"] > coip["id2"]] = coip.loc[coip["id1"] > coip["id2"], ["id2", "id1", "score"]]
coip = coip.groupby(["id1", "id2"], as_index = False).mean()
# Filter 
coip = coip[coip["id1"].isin(uniprot_mapping) & coip["id2"].isin(uniprot_mapping)]
coip.to_csv("../data/networks/coip_full.tsv", sep = "\t", header = None, index = None)
```

```python
# Get the COIP shared network
coip_s = pd.read_csv("../data/networks/coip_full_shared.tsv", sep = "\t", header = None)
# Filter
coip_s.loc[coip_s[0] > coip_s[1]]  = coip_s.loc[coip_s[0] > coip_s[1], [1, 0, 2]]
coip_s = coip_s[coip_s[0].isin(uniprot_mapping) & coip_s[1].isin(uniprot_mapping)]
coip_s = coip_s.groupby([0, 1], as_index = False).mean()
# Save
coip_s.to_csv("../data/networks/coip_full_shared.tsv", sep = "\t", header = None, index = None)
coip_s
```

```python
dfc = pd.read_csv("../data/networks/coip_hc_full.tsv", sep = "\t", header = None)
dfy = pd.read_csv("../data/networks/y2h_hc_full.tsv", sep = "\t", header = None)
```

```python
nodec = set(list(dfc[0]) + list(dfc[1]))
nodey = set(list(dfy[0]) + list(dfy[1]))
nodei = nodec.intersection(nodey)
len(nodec), len(nodey), len(nodei)
dfc_sh = dfc[dfc[0].isin(nodei) & dfc[1].isin(nodei)]
dfy_sh = dfy[dfy[0].isin(nodei) & dfy[1].isin(nodei)]
```

```python
gcsh = nx.from_pandas_edgelist(dfc_sh, 0, 1, 2)
gcc  = max(nx.connected_components(gcsh), key = len)
gcsh_cc = gcsh.subgraph(gcc)
df_cc = nx.to_pandas_edgelist(gcsh_cc)
df_cc
# df_cc.to_csv("../data/networks/coip_hc_shared.tsv", sep = "\t", header = None, index = None)
```

```python
gysh = nx.from_pandas_edgelist(dfy_sh, 0, 1, 2)
ycc = max(nx.connected_components(gysh), key = len)
gysh_cc = gysh.subgraph(ycc)
df_cc = nx.to_pandas_edgelist(gysh_cc)
df_cc.to_csv("../data/networks/y2h_hc_shared.tsv", sep = "\t", header = None, index = None)
```
