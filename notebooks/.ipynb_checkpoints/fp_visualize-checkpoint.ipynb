```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualize Function prediction MV

```python
import pandas as pd

df = pd.read_csv("data+results/mv_function_prediction_results_y2h-coip.tsv", sep = "\t", index_col=0)
df = df.drop(["STD"], axis = 1)
df
```

```python
# Define the palette as a list to specify exact values
palette = sns.color_palette("rocket_r")

df.columns = ["Metric", "k", "GO", "Network", "Score"]
sns.relplot(data = df,
            x    = "k",
            y    = "Score",
            hue  = "Network",
            size = "Metric",
            size_order=["Acc", "F1"],
            col  = "GO",
            kind = "line")
plt.savefig("diagrams/fp_mv_results.svg")
```

## Visualize Function prediction L3

```python
df = pd.read_csv("data+results/l3_function_prediction_results_y2h-coip.tsv", sep = "\t", index_col=0)
df = df.drop(["STD"], axis = 1)
```

```python
df.columns = ["Metric", "k", "GO", "Network", "Score"]
sns.relplot(data = df,
            x    = "k",
            y    = "Score",
            hue  = "Network",
            size = "Metric",
            size_order=["Acc", "F1"],
            col  = "GO",
            kind = "line")
plt.savefig("diagrams/fp_l3_results.png", dpi = 300)
```

```python
dfmv = pd.read_csv("data+results/mv_function_prediction_results_y2h-coip.tsv", sep = "\t", index_col=0)
dfl3 = pd.read_csv("data+results/l3_function_prediction_results_y2h-coip.tsv", sep = "\t", index_col=0)
dfcomb = pd.merge(dfmv, dfl3, how="inner", on=["Metric", "k", "GO", "Network"], suffixes=["-MV", "-L3"])
dfcomb = dfcomb.drop([dfcomb.columns[i] for i in [5, 7]], axis = 1)
dfcomb = dfcomb.melt(id_vars = ["Metric", "k", "GO", "Network"], 
           var_name = "score_type",
           value_name = "score")
out_score = {"AVG-MV": "MV", "AVG-L3" : "L3"}
dfcomb = dfcomb.replace({"score_type": out_score})
dfcomb
```

```python
sns.relplot(data = dfcomb[dfcomb["Metric"] == "Acc"],
            x    = "k",
            y    = "score",
            hue  = "Network",
            size = "score_type",
            size_order=["L3", "MV"],
            col  = "GO",
            col_order=["F", "P", "C"],
            kind = "line")
plt.savefig("diagrams/fp_mv_l3_results.acc.svg")
```

## Degree Visualization MV

```python
df = pd.read_csv("data+results/func_pred_degree_analysis_mv.tsv", sep = "\t")
df = df.drop(["Score-std"], axis = 1)
#df = df[(df["Test Degree Threshold"] <=100) & (df["Train Degree Threshold"] <=100)]
df
```

```python
df_mv =  pd.read_csv("data+results/func_pred_degree_analysis_mv.tsv", sep = "\t").drop(["Score-std"], axis = 1)
df_l3 = pd.read_csv("data+results/func_pred_degree_analysis_l3.tsv", sep = "\t").drop(["Score-Std"], axis = 1)
df = pd.merge(df_mv, df_l3, how = "inner", on = ["Test Degree Threshold", "Train Degree Threshold", "Network", "GO-type", "Metric", "k"], suffixes=["-mv", "-l3"])
df = df.melt(id_vars=["Test Degree Threshold", "Train Degree Threshold",
                     "Network", "GO-type", "Metric", "k"],
            var_name = "method", value_name="Score")
df = df.replace({"method": {"Score-Avg-mv": "MV", "Score-Avg-l3": "L3"}})
df
```

```python
# Set k = 50, metric = Accuracy
dff  = df[(df["k"] == 50)].drop(["k"], axis = 1)
methods = ["MV", "L3"]
gos  = ["MF", "BP", "CC"]

fig, axes = plt.subplots(2, 3, figsize = (15, 10), sharey = True, sharex = True)
fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
plt.xlabel("Test Degree Threshold")
plt.ylabel("Train Degree Threshold")

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

for i, method in enumerate(methods):
    dfc= dff[dff["method"] == method]
    for j, go in enumerate(gos):
        dfy = dfc[(dfc["Network"] == "Shared-Y2H") & (dfc["GO-type"] == go)].drop(["Network", "GO-type"], axis = 1)[["Train Degree Threshold", 
                                                                                                            "Test Degree Threshold", 
                                                                                                            "Score-Avg"]]
        dfyp = dfy.pivot("Train Degree Threshold", "Test Degree Threshold", "Score-Avg")
        
        dfco = dfc[(dfc["Network"] == "Shared-CoIP") & (dfc["GO-type"] == go)].drop(["Network", "GO-type"], axis = 1)[["Train Degree Threshold", 
                                                                                                            "Test Degree Threshold", 
                                                                                                            "Score-Avg"]]
        dfcop = dfco.pivot("Train Degree Threshold", "Test Degree Threshold", "Score-Avg")

        dfdiff = dfcop.subtract(dfyp)
        sns.heatmap(dfdiff, ax = axes[i][j])
        axes[i][j].invert_yaxis()
        axes[i][j].set_xlabel("")
        axes[i][j].set_ylabel("")
        axes[i][j].set_title(f"Metric={metric} | GO={go}")
fig.tight_layout()
plt.show()
fig.savefig("diagrams/difference_mv_degrees.png", dpi = 300)
```
